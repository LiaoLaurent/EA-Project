```python
import numpy as np
from mfg_without_common_noise import plot_Nsteps
from one_step import plot_onestep
```

```python
# Custom Parameters
T = 1.0  # Time horizon
N_time_steps = 1000  # Number of time steps
delta_t = T / N_time_steps  # Time step size
theta = 3  # Mean reversion speed

m_bar = -2  # Mean in drift dynamics
sigma = 0.5  # Initial Distribution std

sigma_bar = sigma * np.sqrt(2 * theta) # Idiosyncratic Risk
```

```python
# Plot N_times_steps trajectories for N_agents Mean Field Game Without Common Noise 
N_agents = 1000
plot_Nsteps(N_agents, N_time_steps, delta_t, theta, m_bar, sigma_bar, T, save_plots=False)
```

```python
# Plot 1 step simulation of agents trajectories
mu = m_bar
plot_onestep(mu, sigma, theta, m_bar, T, save_plots=False)
```
